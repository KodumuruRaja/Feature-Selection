```
import pandas as pd
data=pd.DataFrame({'A':[1,2,3,1,2,3],
                   'B':[4,5,6,7,8,9],
                   'C':[0,0,0,0,0,0],
                   'D':[1,1,1,1,1,1]})
```

```
data.head()
```

# **Drop Constant Features**

```
# It will drop Zero Variance Features
from sklearn.feature_selection import VarianceThreshold
var=VarianceThreshold(threshold=0)
var.fit(data)
```

```
var.get_support()
```

```
data.columns[var.get_support()]
```

```
constant_columns=[column for column in data.columns
                  if column not in data.columns[var.get_support()]]
```

```
len(constant_columns)
```

```
for feature in constant_columns:
  print(feature)
```

```
data.drop(constant_columns,axis=1)
```

```
# Dropping Constant Features in DataSEt
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
```

```
data=pd.read_csv("/content/train.csv")
```

```
data.shape
```

```
data.head()
```

```
x=data.drop(labels=['TARGET'],axis=1)
y=data['TARGET']
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```
x_train.shape,x_test.shape
```

```
# Applying Variance Threshold
var=VarianceThreshold(threshold=0)
var.fit(x_train)
```

```
sum(var.get_support())
```

```
len(x_train.columns[var.get_support()])
```

```
cc=[column for column in x_train.columns if column not in x_train.columns[var.get_support()]]
```

```
len(cc)
```

```
for column in cc:
  print(column)
```

```
x_train.drop(cc,axis=1)
```
